```python
new_rows = []

for i in range(len(df)):
    prompt = df.iloc[i]["prompt"]
    category = "toxicity"
    sub_category = df.iloc[i]["crime"]
    target = (
        df.iloc[i]["region_specific_topic"]
        + " The use cases include the following: "
        + df.iloc[i]["use_case"]
    )
    context = {
        "region_specific_topic": df.iloc[i]["region_specific_topic"],
        "region": df.iloc[i]["region"],
        "why_prompt_tailored_for_region": df.iloc[i]["why_prompt_tailored_for_region"],
        "medium_keyword": df.iloc[i]["medium_keyword"],
        "why_prompt_harmful": df.iloc[i]["why_prompt_harmful"],
        "why_prompt_contains_instruction_keyword": df.iloc[i][
            "why_prompt_contains_instruction_keyword"
        ],
        "use_case": df.iloc[i]["use_case"],
    }
    statistical_analysis = {}

    new_rows.append(
        {
            "prompt": prompt,
            "category": category,
            "sub-category": sub_category,
            "target": target,
            "context": context,
            "statistical-analysis": statistical_analysis,
        }
    )

    # Create a new DataFrame
    preprocessed_df = pd.DataFrame(new_rows)
```
