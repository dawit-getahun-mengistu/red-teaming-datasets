```python
def load_dict(str_df):
    return ast.literal_eval(str_df)


new_rows = []

for i in range(len(df)):
    prompt_data = load_dict(df.iloc[i]["prompt"])
    if not prompt_data["toxicity"]:
        print(prompt_data)
    elif prompt_data["toxicity"] > 0.5:
        cont_data = load_dict(df.iloc[i]["continuation"])
        prompt = prompt_data["text"]
        category = "toxicity"
        sub_category = ""
        target = ""
        context = {
            "prompt": prompt_data["text"],
            "prompt_continuation": cont_data["text"],
        }
        statistical_analysis = {
            "prompt": {k: v for k, v in prompt_data.items() if k != "text"},
            "prompt_continuation": {k: v for k, v in cont_data.items() if k != "text"},
        }
        new_rows.append(
            {
                "prompt": prompt,
                "category": category,
                "sub-category": sub_category,
                "target": target,
                "context": context,
                "statistical-analysis": statistical_analysis,
            }
        )

# Create a new DataFrame
preprocessed_df = pd.DataFrame(new_rows)
```
