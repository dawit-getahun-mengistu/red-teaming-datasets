```python
import pandas as pd
```

```python
en_df = pd.read_csv("data_en.csv")
zh_df = pd.read_csv("data_zh.csv")
```

```python
new_rows = []

for i in range(len(en_df)):
    prompt = en_df.question[i]
    category = en_df.types_of_harm[i]
    sub_category = en_df.risk_area[i] + ", " + en_df.specific_harms[i]
    target = ""
    context = {}
    statistical_analysis = {}

    new_rows.append(
        {
            "prompt": prompt,
            "category": category,
            "sub-category": sub_category,
            "target": target,
            "context": context,
            "statistical-analysis": statistical_analysis,
        }
    )

# Create a new DataFrame
preprocessed_df = pd.DataFrame(new_rows)
```

```python
preprocessed_df.to_csv("16-do-not-answer-en-prompts.csv")
```

```python
new_rows = []

for i in range(len(zh_df)):
    prompt = zh_df.en_question[i]
    category = zh_df.types_of_harm[i]
    sub_category = zh_df.risk_area[i] + ", " + zh_df.specific_harms[i]
    target = ""
    context = {"chinese": zh_df.zh_question[i]}
    statistical_analysis = {}

    new_rows.append(
        {
            "prompt": prompt,
            "category": category,
            "sub-category": sub_category,
            "target": target,
            "context": context,
            "statistical-analysis": statistical_analysis,
        }
    )

    # Chinese
    prompt = zh_df.zh_question[i]
    category = zh_df.types_of_harm[i]
    sub_category = zh_df.risk_area[i] + ", " + zh_df.specific_harms[i]
    target = ""
    context = {"english": zh_df.en_question[i]}
    statistical_analysis = {}

    new_rows.append(
        {
            "prompt": prompt,
            "category": category,
            "sub-category": sub_category,
            "target": target,
            "context": context,
            "statistical-analysis": statistical_analysis,
        }
    )

# Create a new DataFrame
preprocessed_df = pd.DataFrame(new_rows)
```
