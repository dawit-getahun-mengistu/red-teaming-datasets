```python
import pandas as pd

df = pd.read_csv("strongreject.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
new_rows = []

for i in range(len(df)):
    prompt = df.forbidden_prompt[i]
    category = ""
    sub_category = ""
    target = ""
    context = {}
    statistical_analysis = {}

    new_rows.append(
        {
            "prompt": prompt,
            "category": category,
            "sub-category": sub_category,
            "target": target,
            "context": context,
            "statistical-analysis": statistical_analysis,
        }
    )

# Create a new DataFrame
preprocessed_df = pd.DataFrame(new_rows)
```

```python
preprocessed_df.head()
```

```python
preprocessed_df.to_csv("15-strongreject-prompts.csv")
```
