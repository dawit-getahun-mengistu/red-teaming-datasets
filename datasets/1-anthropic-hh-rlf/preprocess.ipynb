```python
new_rows = []

for i in range(len(df)):
    transcript = df.iloc[i]["transcript"]
    target = df.iloc[i]["task_description"]
    category = ""
    context = {
        "task_description": df.iloc[i]["task_description"],
        "source_model_type": df.iloc[i]["model_type"],
        "source_num_params": df.iloc[i]["num_params"],
    }
    statistical_analysis = {
        "min_harmlessness_score_transcript": df.iloc[i][
            "min_harmlessness_score_transcript"
        ],
        "task_descripton_harmlessness_score": df.iloc[i][
            "task_descripton_harmlessness_score"
        ],
        "rating": df.iloc[i]["rating"],
    }
    # new_rows.append([transcript, target, category, context, statistical_analysis])

    new_rows.append(
        {
            "transcript": transcript,
            "category": category,
            "sub-category": category,
            "target": target,
            "context": context,
            "statistical-analysis": statistical_analysis,
        }
    )

# Create new DataFram
preprocessed_df = pd.DataFrame(new_rows)
```
